```python
data_dir = '/content/uncompressed/zalando'
```

```python
import pathlib
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
clothes_images_dict= {
    'hoodies': list(data_dir.glob('hoodies/*.jpg'))[:500],
    'shirt': list(data_dir.glob('shirt/*.jpg'))[:500],
    'longsleeve': list(data_dir.glob('longsleeve/*.jpg'))[:500],
}
```

```python
clothes_labels_dict= {
    'hoodies': 0,
    'shirt': 1,
    'longsleeve': 2,
}
```

```python
x, y = [],[]
for clothes_name, images in clothes_images_dict.items():
  for image in images:
    if image is not None:
      img = cv2.imread(str(image))
      resized_img  = cv2.resize(img,(180,180))
      x.append(resized_img)
      y.append(clothes_labels_dict[clothes_name])

```

```python
x = np.array(x)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
num_classes = 3
```

```python
model = Sequential([
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test, y_test)
```
